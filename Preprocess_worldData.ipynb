# Preprocess Global Food Prices data from the World Food Programme

```python
import os

# Specify the directory name
directory_name = "./data/output"

# Create the directory
try:
    os.mkdir(directory_name)
    print(f"Directory '{directory_name}' created successfully.")
except FileExistsError:
    print(f"Directory '{directory_name}' already exists.")
except PermissionError:
    print(f"Permission denied: Unable to create '{directory_name}'.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import pandas as pd

# Load CSV file into a DataFrame
df = pd.read_csv('./data/wfpvam_foodprices.csv')

# Display first few rows
print(df.head())
```

```python
df.describe()
```

```python
filtered_df = df[df['adm0_name'] == 'Philippines'] # Only Philippine Produce
filtered_df = filtered_df[filtered_df['pt_id'] == 15] # Only Retail Prices
filtered_df = filtered_df.drop(['adm0_id', 'adm0_name', 'adm1_id', 'adm1_name', 'mkt_id', 'mkt_name', 'cur_id', 'cur_name', 'pt_id', 'pt_name', 'mp_commoditysource'], axis=1) # Only Relevant Columns

filtered_df.head()
```

```python
print(filtered_df.shape)
```

```python
values = df['cm_name'].unique()
```

```python
unique_values = df['cm_name'].unique()

print(unique_values.size)

print(unique_values)
```

```python
rice_df = filtered_df[filtered_df['cm_name'].str.contains('rice', case=False)]
rice_df = rice_df[rice_df['cm_name'] == 'Rice (regular, milled) - Retail']
rice_df.describe()

rice_df.to_csv('./data/output/rice.csv', index=False)
```
